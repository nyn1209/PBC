# Narrative Summary: Primary Biliary Cirrhosis Treatment Analysis
This project presents an in-depth analysis of patient data to evaluate the impact of Penicillamine versus Placebo on survival outcomes in individuals diagnosed with Primary Biliary Cirrhosis (PBC) — a chronic liver disease. Utilizing a comprehensive dataset, the project walks through each critical stage of data processing, statistical modeling, and interpretation to extract actionable clinical insights.


### **1. Data Understanding and Cleaning**
The journey began by loading and inspecting the dataset, identifying and resolving key issues like duplicates and missing values. A gender-sensitive imputation strategy was employed to fill gaps in medical indicators such as cholesterol, copper, triglycerides, and platelets, ensuring demographic consistency in preprocessing.

### **2. Exploratory Data Analysis (EDA)**
A combination of univariate, bivariate, and multivariate analysis revealed crucial insights:
- **Gender Bias**: A noticeable skew toward female patients, which aligns with medical literature suggesting higher PBC incidence in women.
- **Outliers**: Indicators like **alkaline phosphatase** and **bilirubin** showed extreme values, strongly correlating with mortality risk.
- **Age and Status**: Older patients had higher mortality, and liver transplant candidates were generally younger, suggesting selective criteria in clinical decision-making.

### **3. Medical Indicators and Survival Outcomes**
Visual comparisons of health indicators across treatment groups indicated that while some parameters like bilirubin and albumin showed variation, there were **no major differences** between penicillamine and placebo groups at a descriptive level.

### **4. Predictive Modeling**
Using **hierarchical logistic regression**, it was found that:
- **Gender**, **bilirubin**, **albumin**, and **protime** were significant predictors of mortality.
- The **type of drug (penicillamine)** was not statistically significant when controlling for other variables.
- The model achieved **AUC = 0.82** and **Accuracy = 78%**, indicating strong predictive power.

### **5. Survival Analysis**
Two approaches—**Kaplan-Meier curves** and **Cox Proportional Hazards Models**—were used:
- Kaplan-Meier suggested **penicillamine may extend survival**, especially in patients aged 30–50.
- However, the Cox model, which controls for disease stage, found **no significant survival benefit** from penicillamine, highlighting the dominant role of **disease stage** in predicting mortality.

### **6. Clustering for Patient Profiling**
Unsupervised learning (KMeans clustering) was used to group patients based on liver health indicators:
- Clusters helped reveal subpopulations with distinct mortality risks.
- Patients with **high bilirubin and low albumin** (Cluster 0) showed the **highest mortality**, underscoring their clinical vulnerability.

### **7. Follow-up Patterns**
Patients in **early disease stages** had significantly longer follow-up durations. This aligns with expectations, as severe-stage patients tend to deteriorate faster, requiring either intervention or resulting in death sooner.

### **Key Insights**
- **Stage of disease** is the strongest predictor of survival—not the treatment type.
- **Penicillamine** may offer marginal survival benefits in earlier stages, but its overall effect is not statistically significant.
- **Gender and liver-related indicators** such as **bilirubin** and **albumin** are important mortality predictors.
- **Clustering analysis** can be a powerful tool for patient risk stratification.

### **Conclusion**
While penicillamine shows some promise in descriptive survival analysis, rigorous modeling suggests **limited independent effect** when controlling for other clinical factors. Future studies should incorporate **dosage levels**, **longitudinal tracking**, and potentially combine treatment with other interventions to better understand therapeutic value.

### Outliers

```python
columns = ['bili', 'chol', 'albumin', 'copper', 'alk.phos', 'sgot', 'trig', 'platelet', 'protime']

def outlier_info(column):
    Q1 = df1[column].quantile(0.25)
    Q3 = df1[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df1[(df1[column] < lower_bound) | (df1[column] > upper_bound)]
    percentage = (len(outliers) / len(df1)) * 100
    return (lower_bound, upper_bound), percentage

outlier_info_dict = {col: outlier_info(col) for col in columns}

for col, (range_values, percentage) in outlier_info_dict.items():
    print(f"Normal range for {col}: {range_values}")
    print(f"Outlier percentage for {col}: {percentage:.2f}%\n")
```

# **4.Hierarchical Logistic Regression Model**

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
```

```python
# Separate patients into 2 groups based on drug type: penicillamine and placebo
df1['drug_binary'] = df1['drug'].apply(lambda x: 1 if x == 'penicillamine' else 0)  # 1 for penicillamine, 0 for placebo
```

```python
# Selecting relevant variables for modeling

df1['sex'] = df1['sex'].astype('category')
df1['sex'] = df1['sex'].cat.codes #female 0, male 1

X = df1[['drug_binary', 'age_years', 'sex', 'bili', 'chol', 'albumin', 'copper', 'alk.phos', 'sgot', 'trig', 'platelet', 'protime']]
X = pd.get_dummies(X, drop_first=True)  # Convert categorical variables to dummy variables
y = df1['status_binary']
```

```python
df1.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X.astype(float), y.astype(float), test_size=0.3, random_state=42)
```

```python
# Fit logistic regression model
logit_model = sm.Logit(y_train, sm.add_constant(X_train))
result = logit_model.fit()
print(result.summary())
```

1. p-value for the `drug_binary` coefficient (0.5683) = 0.154 ➔ not statistically significant (but close to the threshold), indicating no significant difference in mortality risk between those treated with penicillamine and placebo.

2. `Sex` has a significant coefficient (1.1588, p-value = 0.0045): gender is an important factor influencing mortality risk, with men (male=1) having a higher likelihood of mortality than women (although this could be due to sample bias in the dataset).

3. **Health Indicators**:
   - `bilirubin`: coefficient = 0.2221 (p-value = 0.032) ➔ higher levels are associated with an increased mortality risk.
   - `albumin`: coefficient = -0.9451 (p-value = 0.053) ➔ higher albumin levels are associated with a notable reduction in mortality risk.
   - `protime`: coefficient = 0.8974 (p-value = 0.001) ➔ prolonged blood clotting time increases mortality risk.

4. Pseudo R-squared value = 0.3743 ➔ approximately 37.43% of the variance in mortality risk can be explained by the predictors in the model ➔ **moderate fit**.

5. Non-significant predictors:
   - Cholesterol (`chol`), copper, SGOT, triglycerides (`trig`), and platelets (`platelet`) have non-significant p-values ➔ they do not influence mortality risk.





```python
# Predicting the probability of death on the test set
y_pred_prob = result.predict(sm.add_constant(X_test))
```

```python
# Calculate AUC (Area Under the Curve)
from sklearn.metrics import roc_auc_score, accuracy_score

auc = roc_auc_score(y_test, y_pred_prob)
print(f"AUC: {auc}")
```

```python
# Calculate accuracy with threshold of 0.5
y_pred = (y_pred_prob >= 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

AUC is a performance metric used to assess the prediction quality of the model. AUC values range from 0.5 to 1, where 0.5 corresponds to no discrimination ability (random prediction).

➔ **AUC = 0.82, Accuracy score = 0.78** indicates that the model has fairly good predictive accuracy, but there is still about 25% inaccuracy, suggesting room for improvement.

```python
from lifelines import KaplanMeierFitter
```

```python
# Separate the groups by drug type
penicillamine_group = df1[df1['drug'] == 'penicillamine']
placebo_group = df1[df1['drug'] == 'placebo']

# Create Kaplan-Meier Fitter instance
kmf = KaplanMeierFitter()

plt.figure(figsize=(10, 6))

# Fit the model for penicillamine group
kmf.fit(penicillamine_group['age_years'], event_observed=penicillamine_group['status_binary'], label='Penicillamine')
kmf.plot(ci_show=False, color='skyblue')

# Fit the model for placebo group
kmf.fit(placebo_group['age_years'], event_observed=placebo_group['status_binary'], label='Placebo')
kmf.plot(ci_show=False, color='lightsalmon')

plt.title('Kaplan-Meier survival curve by drug type')
plt.xlabel('Age (years)')
plt.ylabel('Survival probability')
plt.show()
```

**Key observation:**

1. Higher survival rate with penicillamine suggests that penicillamine may be more effective in prolonging patient survival.
2. The placebo group shows a faster decline in survival after about age 30, whereas the penicillamine group maintains a higher survival rate up to around age 50. ➔ **Penicillamine significantly reduces mortality risk compared to placebo, especially in older age groups.**
3. In the initial stages, both survival curves are close to 1 due to the age range in the dataset starting from middle age and up.



## Logistic regression and Kaplan-Meier
Logistic regression and Kaplan-Meier show different results regarding the effect of drug type.

* Kaplan-Meier Survival Curve provides non-parametric analysis, examining only one factor (drug type), which clearly shows the survival benefit of penicillamine.
* Logistic regression, adjusted for multiple variables (e.g., age, health indices) and modeling a single outcome (e.g., survival status at the end of the study), shows that drug may not necessarily be an independent predictor of survival when other factors are considered.

# **7.Survival analysis by `fu.days`: Kaplan-Meier and Cox proportional**

In the previous analysis, the decreasing trend in follow-up time may reflect the worsening condition as the disease progresses, with similar follow-up times for both treatment methods showing that `penicillamine` may not provide a significant survival advantage over `placebo`, especially in the later stages.

<br> The Kaplan-Meier and Cox models will help determine if these factors impact each other.


```python
kmf.fit(penicillamine_group['fu.days'], event_observed=penicillamine_group['status_binary'], label='Penicillamine')
kmf.plot(ci_show=False, color='skyblue')

# Fit and plot the Kaplan-Meier curve for the placebo group
kmf.fit(placebo_group['fu.days'], event_observed=placebo_group['status_binary'], label='Placebo')
kmf.plot(ci_show=False, color='lightsalmon')

# Customize plot
plt.title('Kaplan-Meier Survival Curve by Treatment Type')
plt.xlabel('Follow-up Days')
plt.ylabel('Survival Probability')
plt.legend()
plt.grid(True)
plt.show()
```

The difference between the two lines suggests that `penicillamine` may be more effective in prolonging survival than `placebo`. However, the difference is not as pronounced, suggesting that the effect size may not be very large.

```python
from lifelines import CoxPHFitter

cox_df = df1[['stage', 'drug', 'fu.days', 'status_binary']]
cox_df = pd.get_dummies(cox_df, columns=['drug'], drop_first=True)

# Fit Cox proportional hazards model
cph = CoxPHFitter()
cph.fit(cox_df, duration_col='fu.days', event_col='status_binary')
cph.print_summary()
```

1. **Coefficients** `coef`
   - `stage`: 0.82 ➔ A positive coefficient indicates that later stages are associated with a higher mortality risk.
   - `drug_placebo`: -0.15 ➔ Using placebo compared to penicillamine results in a slight decrease in mortality risk, but this coefficient is small and negative.

2. **Hazard Ratios** `exp(coef)`
   - `stage`: 2.28 ➔ Patients in later stages have a 2.28 times higher mortality risk compared to those in early stages.
   - `drug_placebo`: 0.86 ➔ Using placebo instead of penicillamine reduces the mortality risk by approximately 14%. However, this effect may not be statistically significant (confidence interval exceeds 1).

3. **Confidence Intervals** (95% `exp(coef) lower` and 95% `exp(coef) upper`)
   - `stage`: 1.79 - 2.9 ➔ The increased risk in later stages is statistically significant (confidence interval does not exceed 1).
   - `drug_placebo`: 0.61 - 1.23 ➔ Exceeds 1, indicating that the effect of using placebo is not statistically significant.

4. **p-value** (`p`)
   - < 0.05: Statistically significant ➔ No significant difference in mortality risk between those using penicillamine and placebo.

   **Summary**: Disease progression (`stage`) is a strong predictor of mortality risk, while treatment with drugs (placebo compared to penicillamine) does not significantly impact survival outcomes in this analysis.
